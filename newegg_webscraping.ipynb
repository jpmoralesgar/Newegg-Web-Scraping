```python
from bs4 import BeautifulSoup
import requests
import pandas as pd 
from datetime import datetime

full_list_dictionary = []

names_list = []
prices_list = []
cents_list = []
image_links_list = []

for x in range(1,11):

    # HTTP Request
    
    url_filtered = 'https://www.newegg.com/p/pl?N=100007709%20601408872%20601357282%20601359511&PageSize=96&page='

    # url_all = 'https://www.newegg.com/GPUs-Video-Graphics-Cards/SubCategory/ID-48/Page-'+str(x)+'?Tid=7709'

    # Get Request

    response = requests.get(url_filtered+str(x))
    
    # Status Code

    response.status_code

    # Soup Object

    soup = BeautifulSoup(response.content, 'html.parser')

    results = soup.find('div',{'class':'item-cells-wrap border-cells items-grid-view four-cells expulsion-one-cell'})

    # Graphic Card Name , Price and Image links

    names = []
    prices= []
    cents = []
    image_links = []

    for items in results.find_all('div',{'class':'item-container'}):
        name = items.find('a',{'class':'item-title'}).text
        price = str(getattr(items.find('ul',{'class':'price'}).find('strong'),'text', 'none'))
        cent = str(getattr(items.find('sup'),'text','none'))
        image_link = items.find('img').attrs['src']

        names.append(name)
        prices.append(price)
        cents.append(cent)
        image_links.append(image_link)
        
        names_list.append(names)
        prices_list.append(prices)
        cents_list.append(cents)
        image_links_list.append(image_links) 
        
        items_info = {
            'Name': name,
            'price': price,
            'cents': cent,
            'Image_link' : image_link
        }
        
        full_list_dictionary.append(items_info)
        
fecha = datetime.now()

fecha_list =[fecha.strftime("%d/%m/%y")]*len(full_list_dictionary)

# len(fecha_list)
#len(full_list_dictionary)

final_df = pd.json_normalize(full_list_dictionary)
final_df['Total_Price'] = final_df['price'] + final_df['cents']
final_df = final_df.drop(columns = ['price','cents'])
final_df['Fecha'] = fecha_list


from pathlib import Path
fecha_str = str(fecha.strftime("%d/%m/%y"))
fecha_str = fecha_str.replace('/','-')
final_df.to_csv(fecha_str+' '+'Graphic_Cards_csv')

final_df

```
